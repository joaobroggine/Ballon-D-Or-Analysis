```
# RM:557129 João Vitor Broggine Lopes 1TDSPF
# RM:554727 João Victor Rocha Cândido 1TDSPF

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Carregamento e visualização inicial dos dados
print("\nPasso 1: Carregamento e visualização inicial dos dados.")
print("Estamos carregando os dados dos indicados à Bola de Ouro 2024 e vamos dar uma olhada nas primeiras linhas.")
# Carregando o dataset
df = pd.read_csv('sample_data/2024 Ballon Dor Nominees League Stats.csv')

# Visualizando as primeiras linhas do dataset
print("Primeiras linhas do dataset:")
print(df.head())

# Verificando a estrutura do dataset
print("\nAgora vamos ver mais detalhes sobre o dataset, como o tipo de dados e o total de colunas e linhas.")
print(df.info())

# 2. Tratamento de dados faltantes
print("\nPasso 2: Tratamento de dados faltantes.")
print("Verificamos se há informações faltando. Por exemplo, se não soubermos a liga de algum jogador,")
print("não conseguiremos incluir essa informação na análise. Por isso, vamos remover essas linhas.")
# Verificando dados faltantes
print("\nDados faltantes por coluna:")
print(df.isnull().sum())

# Removendo linhas com dados faltantes na coluna "league"
df = df.dropna(subset=['league'])

# 3. Exploração descritiva dos dados
print("\nPasso 3: Análise descritiva.")
print("Agora vamos ver a quantidade de indicados por liga. Isso vai nos ajudar a ver quais ligas")
print("têm mais jogadores concorrendo à Bola de Ouro.")
# Descrição das variáveis categóricas
print("\nDistribuição de indicados por liga:")
print(df['league'].value_counts())

# Estatísticas descritivas para outras colunas numéricas, se aplicável
print("\nVamos também ver algumas estatísticas básicas, como as médias e os valores mínimos e máximos,")
print("caso o dataset tenha mais dados numéricos que sejam interessantes para a análise.")
print(df.describe())

# 4. Criação de visualizações
print("\nPasso 4: Criação de gráficos.")
print("Agora vamos criar um gráfico que mostra quantos jogadores de cada liga foram indicados.")
print("No gráfico, quanto maior a barra, mais jogadores daquela liga foram indicados.")
# Contagem de indicados por liga
plt.figure(figsize=(10,6))
sns.countplot(y='league', data=df, order=df['league'].value_counts().index)
plt.title('Distribuição de Indicados à Bola de Ouro 2024 por Liga')
plt.xlabel('Número de Indicados')
plt.ylabel('Liga')

# Adicionando contagem em cada barra
for i in plt.gca().patches:
    plt.text(i.get_width()+0.3, i.get_y()+0.5, str(int(i.get_width())), fontsize=12)

plt.show()

print("\nExplicação do gráfico:")
print("Este gráfico mostra a quantidade de jogadores de cada liga que foram indicados.")
print("Cada barra representa uma liga, e o comprimento da barra mostra quantos jogadores foram indicados.")
print("Por exemplo, se a barra da 'Premier League' for a mais longa, isso significa que essa liga tem mais indicados.")

# 5. Resposta à questão levantada
print("\nPasso 5: Resposta à questão levantada.")
print("Agora precisamos verificar qual é a liga com o maior número de indicados.")
# Verificando a liga com o maior número de indicados
liga_com_mais_indicados = df['league'].value_counts().idxmax()
print(f"\nCom isso a liga com o maior número de indicados é: {liga_com_mais_indicados}")

# Conclusão
print("\nConclusão:")
print(f"Após analisar os dados, descobrimos que a liga com o maior número de indicados é a {liga_com_mais_indicados}.")
print("Isso mostra que essa liga tem muitos jogadores de destaque no futebol internacional em 2024.")
```
